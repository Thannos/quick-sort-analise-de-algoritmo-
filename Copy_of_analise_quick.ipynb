```
import random
import matplotlib.pyplot as plt
import numpy as np
import time
```

```
class Quick(object):
    def particao(self, a, ini, fim):
        pivo = a[fim-1]
        start = ini
        end = ini
        for i in range(ini,fim):
            if a[i] > pivo:
                end += 1
            else:
                end += 1       
                start += 1
                aux = a[start-1]
                a[start-1] = a[i]
                a[i] = aux
        return start-1
        
    def quickSort(self, a, ini, fim):
        if ini < fim:
            pp = self.randparticao(a, ini, fim)
            #pp = self.particao(a, ini, fim)
            self.quickSort(a, ini, pp)
            self.quickSort(a, pp+1,fim)
        return a
        
    def randparticao(self,a,ini,fim):
        rand = random.randrange(ini,fim)
        aux = a[fim-1]
        a[fim-1] = a[rand]
        a[rand] = aux
        return self.particao(a,ini,fim)

def desc(vet, tam):
  #int aux
  for i in range(0, tam-1):
    for j in range(i, tam):
      if(vet[j]>vet[i]):
        aux=vet[j]
        vet[j]=vet[i]
        vet[i]=aux
```

```
def ale(m,k):
  vetortempo=[0]*k
  media=0
  tempo=0
  for i in range (0,k):
    a = random.sample(range(1,1000), m)
    q = Quick()
    #print (a)
    inicio=time.time()
    #print (q.quickSort(a,0,len(a)))
    q.quickSort(a,0,len(a))
    fim=time.time()
    tempo+=(fim-inicio)
    media=i
    vetortempo[i]=tempo
  tempo/=media
  print('média de execução aleatória: ', round(tempo,5))
  return vetortempo
```

```
def cresc(m,k):
  media=0
  tempo=0
  vetortempo=[0]*k
  for i in range (0,k):
    a = random.sample(range(1,1000), m) 
    q = Quick()
    q.quickSort(a,0,len(a))
    #print (a)
    inicio=time.time()
    #print (q.quickSort(a,0,len(a)))
    q.quickSort(a,0,len(a))
    fim=time.time()
    tempo+=(fim-inicio)
    media=i
    vetortempo[i]=tempo
  tempo/=media
  print('média de execução crescente: ', round(tempo,5))
  return vetortempo
```

```
def decresc(m,k):
  media=0
  tempo=0
  vetortempo=[0]*k
  for i in range (0,k):
    a = random.sample(range(1,1000), m) 
    q = Quick()
    desc(a,m)
    #print (a)
    inicio=time.time()
   # print (q.quickSort(a,0,len(a)))
    q.quickSort(a,0,len(a))
    fim=time.time()
    tempo+=(fim-inicio)
    media=i
    vetortempo[i]=tempo
  tempo/=media
  print('média de execução crescente: ', round(tempo,5))
  return vetortempo
```

```
m=100
k=100
print('100 posições\n')
t1=ale(m,k)
t2=cresc(m,k)
t3=decresc(m,k)
print('\n500 posições\n')
m=500
t11=ale(m,k)
t21=cresc(m,k)
t31=decresc(m,k)
print('\n900 posições\n')
m=900
t12=ale(m,k)
t22=cresc(m,k)
t32=decresc(m,k)
```
